# Energy difference plotting function

```python
def DiffPlot(N,lr,diff_8,diff_16,diff_24,diff_32,diff_40,xlim,ylim):
    'plots the energy differences for each nh'
    
    fig = plt.figure(1,figsize=(6,4), dpi=120, facecolor='w', edgecolor='k')
     
    plt.plot(diff_8,marker='o',markersize=1,linewidth=0.0,markevery=1,label= 'nh = 8')
    plt.plot(diff_16,marker='o',markersize=1,linewidth=0.0,markevery=1,label= 'nh = 16')
    plt.plot(diff_24,marker='o',markersize=1,linewidth=0.0,markevery=1,label= 'nh = 24')
    plt.plot(diff_32,marker='o',markersize=1,linewidth=0.0,markevery=1,label= 'nh = 32')
    plt.plot(diff_40,marker='o',markersize=1,linewidth=0.0,markevery=1,label= 'nh = 40')
    
    plt.xlabel("Step",fontsize=15)
    plt.ylabel("$|\\Delta \\langle H \\rangle|$",fontsize=20)
    plt.title("{} x {} Lattice, Learning Rate = {}".format(N,N,lr))
    plt.legend(loc="best")
    plt.xlim(xlim[0],xlim[1])
    plt.ylim(ylim[0],ylim[1])

    plt.show()
```
