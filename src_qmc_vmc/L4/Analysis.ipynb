```python
import numpy as np
import matplotlib.pyplot as plt
import random

#L=4
exact_energy = -0.4534132086591546
```

```python
a_file = open("E4vmc.dat", "r")
vmc_energy = np.loadtxt(a_file)
a_file.close()

a_file = open("E4qmc_vmc.dat", "r")
qmcvmc_energy = np.loadtxt(a_file)
a_file.close()
```

```python
fig = plt.figure(1,figsize=(6,4), dpi=120, facecolor='w', edgecolor='k')
 
plt.plot(vmc_energy,marker='o',markersize=2,linewidth=0.0,markevery=1,label="VMC")
plt.plot(qmcvmc_energy,marker='s',markersize=2,linewidth=0.0,markevery=1,label="QMC_VMC")

plt.hlines(exact_energy,0, 1000 ,linestyle="--",label="Exact")  


plt.xlabel("Step",fontsize=15)
plt.ylabel("$\\langle H \\rangle$",fontsize=20)
#plt.title("{} sites".format(N))
plt.legend(loc="best")
#plt.xlim(200,500)
plt.ylim(-0.5,-0.2)

plt.show()

```
