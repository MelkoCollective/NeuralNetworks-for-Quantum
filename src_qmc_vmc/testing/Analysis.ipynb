```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
energy_baseline = np.loadtxt('./baseline/E_VMC_4_8.dat')
steps = np.arange(1,len(energy_baseline)+1,1)

E_QMC = np.loadtxt('./test1/E_QMC_training_4_8.dat')
E_VMC = np.loadtxt('./test1/E_VMC_training_4_8.dat')
energy_test1 = np.concatenate((E_QMC,E_VMC),axis = 0)

E_1 = np.loadtxt('./test2/E_QMC_halfVMC_4_8.dat')
E_2 = np.loadtxt('./test2/E_VMCcont_4_8.dat')
energy_test2 = np.concatenate((E_1,E_2),axis=0)
```

```python
plt.plot(steps,energy_baseline,'g',label = 'baseline')
plt.plot(steps,energy_test1,'b',label = 'test1')
plt.plot(steps,energy_test2,'r',label = 'test2')
plt.legend()
plt.ylim(-0.4,-0.3)
plt.xlim(99,200)
plt.show()
```
