```python
import numpy as np
import matplotlib.pyplot as plt
import random

#L=11
exact_energy = -3.1721148735413904
```

```python
def FigPlot(qmc_energy, vmc_energy, qmc_vmc_energy):

    fig = plt.figure(1,figsize=(6,4), dpi=120, facecolor='w', edgecolor='k')
 
    plt.plot(vmc_energy,marker='o',markersize=2,linewidth=0.0,markevery=1,label="VMC, lr=0.01")
    plt.plot(qmc_energy,marker='s',markersize=2,linewidth=0.0,markevery=1,label="QMC, lr=0.01")
    plt.plot(qmc_vmc_energy,marker='^',markersize=2,linewidth=0.0,markevery=1,label="QMC and VMC, lr=0.01")

    plt.hlines(exact_energy,0, 4000 ,linestyle="--",label="Exact")  
    plt.vlines(200,-3.18,-3,linestyle="--",color='r',label="QMC VMC transition")


    plt.xlabel("Step",fontsize=15)
    plt.ylabel("$\\langle H \\rangle$",fontsize=20)
    #plt.title("{} sites".format(N))
    plt.legend(loc="best")
    #plt.xlim(200,500)
    #plt.ylim(-3.2,-3.05)

    plt.show()
```

```python
a_file = open("nh8/QMC_Full/E_TFIM_QMC_11_8.dat", "r")
qmc_1 = np.loadtxt(a_file)
a_file.close()

a_file = open("nh8/VMC_Full/E_TFIM_VMC_11_8.dat", "r")
vmc_1 = np.loadtxt(a_file)
a_file.close()

a_file = open("nh8/QMC_VMC/E_TFIM_QMC_VMC_11_8.dat", "r")
qmc_vmc_1 = np.loadtxt(a_file)
a_file.close()

FigPlot(qmc_1, vmc_1, qmc_vmc_1)
```

```python
def VMCPlot(vmc_1, vmc_2, vmc_3, vmc_4, vmc_5):

    fig = plt.figure(1,figsize=(6,4), dpi=120, facecolor='w', edgecolor='k')
 
    plt.plot(vmc_1,marker='o',markersize=2,linewidth=0.0,markevery=1,label="VMC, nh=8")
    plt.plot(vmc_2,marker='o',markersize=2,linewidth=0.0,markevery=1,label="VMC, nh=16")
    plt.plot(vmc_3,marker='o',markersize=2,linewidth=0.0,markevery=1,label="VMC, nh=24")
    plt.plot(vmc_4,marker='o',markersize=2,linewidth=0.0,markevery=1,label="VMC, nh=32")
    plt.plot(vmc_5,marker='o',markersize=2,linewidth=0.0,markevery=1,label="VMC, nh=32")

    plt.hlines(exact_energy,0, 4000 ,linestyle="--",label="Exact")  

    plt.xlabel("Step",fontsize=15)
    plt.ylabel("$\\langle H \\rangle$",fontsize=20)
    #plt.title("{} sites".format(N))
    plt.legend(loc="best")
    #plt.xlim(200,500)
    #plt.ylim(-3.2,-3.05)

    plt.show()
```

```python
VMCPlot(vmc_1, vmc_2, vmc_3, vmc_4, vmc_5)
```
