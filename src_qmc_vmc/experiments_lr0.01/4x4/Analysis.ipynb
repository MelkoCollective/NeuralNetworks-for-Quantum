```python
import numpy as np
import matplotlib.pyplot as plt
import random

#L=4
exact_energy = -0.4534132086591546
```

```python
def FigPlot(qmc_energy,vmc_energy,qmc_energy_small,vmc_energy_small):

    fig = plt.figure(1,figsize=(6,4), dpi=120, facecolor='w', edgecolor='k')
 
    plt.plot(vmc_energy,marker='o',markersize=2,linewidth=0.0,markevery=1,label="VMC, lr=0.01")
    plt.plot(qmc_energy,marker='s',markersize=2,linewidth=0.0,markevery=1,label="QMC, lr=0.01")
    plt.plot(vmc_energy_small,marker='o',markersize=2,linewidth=0.0,markevery=1,label="VMC, lr=0.001")
    plt.plot(qmc_energy_small,marker='s',markersize=2,linewidth=0.0,markevery=1,label="QMC, lr=0.001")

    plt.hlines(exact_energy,0, 1000 ,linestyle="--",label="Exact")  

    plt.xlabel("Step",fontsize=15)
    plt.ylabel("$\\langle H \\rangle$",fontsize=20)
    #plt.title("{} sites".format(N))
    plt.legend(loc="best")
    #plt.xlim(200,500)
    plt.ylim(-0.5,-0.2)

    plt.show()
```

```python
a_file = open("nh8/QMC_Full/E_QMC_4_8.dat", "r")
qmc_1 = np.loadtxt(a_file)
a_file.close()

a_file = open("nh8/VMC_Full/E_VMC_4_8.dat", "r")
vmc_1 = np.loadtxt(a_file)
a_file.close()

# load smaller learning rate data for comparison
a_file = open("../../experiments/4x4/nh8/QMC_Full/E_QMC_4_8.dat", "r")
qmc_1_small = np.loadtxt(a_file)
a_file.close()

a_file = open("../../experiments/4x4/nh8/VMC_Full/E_VMC_4_8.dat", "r")
vmc_1_small = np.loadtxt(a_file)
a_file.close()

FigPlot(qmc_1,vmc_1,qmc_1_small,vmc_1_small)
```
