```python
import numpy as np
import matplotlib.pyplot as plt
import random

#L=16
exact_energy = -0.380514770608724
```

```python
def FigPlot(qmc_energy,vmc_energy,qmc_vmc_energy,qmc_energy_small,vmc_energy_small,qmc_vmc_energy_small):

    fig = plt.figure(1,figsize=(6,4), dpi=120, facecolor='w', edgecolor='k')
 
    #plt.plot(vmc_energy,marker='o',markersize=2,linewidth=0.0,markevery=1,label="VMC, lr=0.01")
    #plt.plot(qmc_energy,marker='s',markersize=2,linewidth=0.0,markevery=1,label="QMC, lr=0.01")
    #plt.plot(qmc_vmc_energy,marker='^',markersize=2,linewidth=0.0,markevery=1,label="QMC and VMC, lr=0.01")
    plt.plot(vmc_energy_small,marker='o',markersize=2,linewidth=0.0,markevery=1,label="VMC, lr=0.001")
    plt.plot(qmc_energy_small,marker='s',markersize=2,linewidth=0.0,markevery=1,label="QMC, lr=0.001")
    plt.plot(qmc_vmc_energy_small,marker='s',markersize=2,linewidth=0.0,markevery=1,label="QMC and VMC, lr=0.01")

    plt.hlines(exact_energy,0, 4000 ,linestyle="--",label="Exact") 
    plt.vlines(200,-0.5,-0.2,linestyle="--",color='r',label="QMC VMC transition")

    plt.xlabel("Step",fontsize=15)
    plt.ylabel("$\\langle H \\rangle$",fontsize=20)
    #plt.title("{} sites".format(N))
    plt.legend(loc="best")
    #plt.xlim(200,500)
    plt.ylim(-0.5,-0.2)

    plt.show()
```

```python
def FigPlot_2(qmc_energy,vmc_energy,qmc_vmc_energy):

    fig = plt.figure(1,figsize=(6,4), dpi=120, facecolor='w', edgecolor='k')
 
    plt.plot(vmc_energy,marker='o',markersize=2,linewidth=0.0,markevery=1,label="VMC, lr=0.01")
    plt.plot(qmc_energy,marker='s',markersize=2,linewidth=0.0,markevery=1,label="QMC, lr=0.01")
    plt.plot(qmc_vmc_energy,marker='^',markersize=2,linewidth=0.0,markevery=1,label="QMC and VMC, lr=0.01")

    plt.hlines(exact_energy,0, 4000 ,linestyle="--",label="Exact")      
    plt.vlines(200,-0.5,-0.2,linestyle="--",color='r',label="QMC VMC transition")

    plt.xlabel("Step",fontsize=15)
    plt.ylabel("$\\langle H \\rangle$",fontsize=20)
    #plt.title("{} sites".format(N))
    plt.legend(loc="best")
    #plt.xlim(200,500)
    plt.ylim(-0.5,-0.2)

    plt.show()
```
