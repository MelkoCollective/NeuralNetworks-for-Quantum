```python
import numpy as np
import matplotlib.pyplot as plt
import random

#L=12
exact_energy = -0.3884864748124427
```

```python
def FigPlot(qmc_energy,vmc_energy,qmc_vmc_energy):

    fig = plt.figure(1,figsize=(6,4), dpi=120, facecolor='w', edgecolor='k')
 
    plt.plot(vmc_energy,marker='o',markersize=2,linewidth=0.0,markevery=1,label="VMC")
    plt.plot(qmc_energy,marker='s',markersize=2,linewidth=0.0,markevery=1,label="QMC")
    plt.plot(qmc_vmc_energy,marker='^',markersize=2,linewidth=0.0,markevery=1,label="200 epochs QMC, 3800 epochs QMC")

    plt.hlines(exact_energy,0, 4000 ,linestyle="--",label="Exact")  
    plt.vlines(200,-0.4,-0.1,linestyle="--",color='r',label="QMC VMC transition")


    plt.xlabel("Step",fontsize=15)
    plt.ylabel("$\\langle H \\rangle$",fontsize=20)
    #plt.title("{} sites".format(N))
    plt.legend(loc="best")
    plt.ylim(-0.41,-0.08)

    plt.show()
```

# Plots

## 12x12, nh=8

```python
a_file = open("nh8/QMC_Full/E_QMC_12_8.dat", "r")
qmc_1 = np.loadtxt(a_file)
a_file.close()

a_file = open("nh8/VMC_Full/E_VMC_12_8.dat", "r")
vmc_1 = np.loadtxt(a_file)
a_file.close()

a_file = open("nh8/QMC_VMC/E_QMC_VMC_12_8.dat", "r")
qmc_vmc_1 = np.loadtxt(a_file)
a_file.close()

FigPlot(qmc_1,vmc_1,qmc_vmc_1)
```
