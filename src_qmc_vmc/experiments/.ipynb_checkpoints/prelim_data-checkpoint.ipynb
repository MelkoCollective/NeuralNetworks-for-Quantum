### Plotting

```python
def plot_E(x, energies_QMC, energies_VMC, exact_energy, dim, total_epochs, cutoff, x_lim, y_lim):
  '''
  Plots the energy over epochs. If a hybrid approach was used, the cutoff shows the step at which 
  the approaches changes. If the cutoff is zero, either VMC or QMC was used the whole time.
  In this notebook, we are comparing only QMC and VMC.
  '''
  N = dim ** 2
  fig = plt.figure(1,figsize=(6,2.5), dpi=120, facecolor='w', edgecolor='k')
  plt.plot(x,energies_QMC,marker='o',markersize=1,linewidth=0.0,markevery=1,label= 'QMC')
  plt.plot(x,energies_VMC,marker='o',markersize=1,linewidth=0.0,markevery=1,label= 'VMC')
  plt.hlines(exact_energy,0, total_epochs,linestyle="--",label="Exact")
  if cutoff != 0:
    plt.vlines(cutoff, energy[0], exact_energy, linestyle="--", label="QMC train cutoff")
  plt.xlabel("Step",fontsize=15)
  plt.ylabel("$\\langle H \\rangle$",fontsize=20)
  plt.title("{} sites".format(N))
  plt.legend(loc="best")
  if x_lim[0] != 0 or x_lim[1] != 0:
    plt.xlim(x_lim[0],x_lim[1])
  if y_lim[0] != 0 or y_lim[1] != 0:
    plt.ylim(y_lim[0],y_lim[1])
  plt.show()
```
