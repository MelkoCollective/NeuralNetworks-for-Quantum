```python
import numpy as np
import matplotlib.pyplot as plt
import random

#L=8
exact_energy = -0.40518005298872917
```

```python
def FigPlot(qmc_energy,vmc_energy,qmc_vmc_energy):

    fig = plt.figure(1,figsize=(6,4), dpi=120, facecolor='w', edgecolor='k')
 
    plt.plot(vmc_energy,marker='o',markersize=2,linewidth=0.0,markevery=1,label="VMC")
    plt.plot(qmc_energy,marker='s',markersize=2,linewidth=0.0,markevery=1,label="QMC")
    plt.plot(qmc_vmc_energy,marker='^',markersize=2,linewidth=0.0,markevery=1,label="200 epochs QMC, 3800 epochs VMC")

    plt.hlines(exact_energy,0, 4000 ,linestyle="--",label="Exact")  
    plt.vlines(200,-0.5,-0.15,linestyle="--",color='red',label="QMC to VMC transition")


    plt.xlabel("Step",fontsize=15)
    plt.ylabel("$\\langle H \\rangle$",fontsize=20)
    #plt.title("{} sites".format(N))
    plt.legend(loc="best")
    #plt.xlim(200,500)
    plt.ylim(-0.42,-0.12)

    plt.show()
```

```python
a_file = open("nh8/QMC_Full/E_QMC_8_8.dat", "r")
qmc_1 = np.loadtxt(a_file)
a_file.close()

a_file = open("nh8/VMC_Full/E_VMC_8_8.dat", "r")
vmc_1 = np.loadtxt(a_file)
a_file.close()

a_file = open("nh8/QMC_VMC/E_QMC_VMC_8_8.dat", "r")
qmc_vmc_1 = np.loadtxt(a_file)
a_file.close()

FigPlot(qmc_1,vmc_1,qmc_vmc_1)
```
