# Energy Plotting Function

```python
def QMCPlot(N,lr,exactE,qmc_8,qmc_16,qmc_24,qmc_32,qmc_40,xlim,ylim):

    fig = plt.figure(1,figsize=(6,4), dpi=120, facecolor='w', edgecolor='k')
    
    plt.plot(qmc_8,marker='o',markersize=1,linewidth=0.0,markevery=1,label="nh = 8")
    plt.plot(qmc_16,marker='o',markersize=1,linewidth=0.0,markevery=1,label="nh = 16")
    plt.plot(qmc_24,marker='o',markersize=1,linewidth=0.0,markevery=1,label="nh = 24")
    plt.plot(qmc_32,marker='o',markersize=1,linewidth=0.0,markevery=1,label="nh = 32")
    plt.plot(qmc_40,marker='o',markersize=1,linewidth=0.0,markevery=1,label="nh = 40")

    plt.hlines(exactE,0, 4000 ,linestyle="--",label="Exact")

    plt.xlabel("Step",fontsize=15)
    plt.ylabel("$\\langle H \\rangle$",fontsize=20)
    plt.title("{} x {} Lattice, Learning Rate = {}".format(N,N,lr))
    plt.legend(loc="best")
    plt.xlim(xlim[0],xlim[1])
    plt.ylim(ylim[0],ylim[1])

    plt.show()
```

# Find the moving average for the QMC datasets

```python
def MovingAvg(dataset,n):
    
    L = len(dataset)
    averages = np.zeros(L)
    
    for i in range(n,L-n):
        sum = dataset[i]
        for j in range(n):
            sum += dataset[i+j]
            sum += dataset[i-j]
        averages[i] = sum/(2*n+1)
    
    return averages[n:L-n]
```

# Minima Plotting Functions

```python
def MinPlot1(N,lr,QMC_mins,exactE):
    'plots the minima'
    
    fig = plt.figure(1,figsize=(6,4), dpi=120, facecolor='w', edgecolor='k')
    
    plt.hlines(exactE,0, QMC_mins[-1,0] ,linestyle="--",label="Exact")  
    plt.plot(QMC_mins[:,0],QMC_mins[:,1],marker='o',markersize=3,linewidth=0.0,markevery=1)
    plt.xlabel("# hidden units",fontsize=15)
    plt.ylabel("minimum (average) energy",fontsize=15)
    plt.title("{} x {} Lattice, Learning Rate = {}".format(N,N,lr))

    plt.show()
    
def MinPlot2(N,lr,QMC_minstep):
    'plots the step where minimum found'
    
    fig = plt.figure(1,figsize=(6,4), dpi=120, facecolor='w', edgecolor='k')
    
    plt.plot(QMC_minstep[:,0],QMC_minstep[:,1],marker='o',markersize=3,linewidth=0.0,markevery=1)
    plt.xlabel("# hidden units",fontsize=15)
    plt.ylabel("step where min found",fontsize=15)
    plt.title("{} x {} Lattice, Learning Rate = {}".format(N,N,lr))

    plt.show()
```
